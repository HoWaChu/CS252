```python
import matplotlib.pyplot as plt
arr2 = np.genfromtxt('data/lab1.csv', delimiter = ',')
print(arr2)
plt.figure(figsize=(12,8))
plt.plot(arr2[:, 2], arr2[:, 1], 'r--')
plt.xlabel("Column 2")
plt.ylabel("Column 1")
plt.title("Column 2 vs. Column 1")
```

```python
plt.subplots_adjust(top = 3, right = 5)

plt.subplot(3,2,1)
plt.plot(arr2[:, 1], arr2[:, 0], 'r--')

plt.subplot(3,2,2)
plt.scatter(arr2[:, 1], arr2[:, 0])

plt.subplot(3,2,3)
plt.plot(arr2[:, 2], arr2[:, 1], 'r--')

plt.subplot(3,2,4)
plt.scatter(arr2[:, 2], arr2[:, 1])

plt.subplot(3,2,5)
plt.plot(arr2[:, 2], arr2[:, 0], 'r--')

plt.subplot(3,2,6)
plt.scatter(arr2[:, 2], arr2[:, 0])
```

The second row of graphs are easier to read as a line and as a scatter plot. This is because the data is shaped like a sine graph. 

However, for scatter plots like the first and third row, without the line graph, it's very difficult to see the connection of points.
